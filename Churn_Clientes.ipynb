# Gráifcos de intéres

## 1) Histogramas

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo general
sns.set(style="whitegrid")

# Creo los histogramas en fila
fig, axes = plt.subplots(1, 2, figsize=(18, 5))

# Histograma 1: Meses de suscripción
sns.histplot(data=df_processed, x='Meses de suscripción', kde=True, bins=20, ax=axes[0], color='skyblue')
axes[0].set_title('Distribución de Meses de Suscripción')

# Histograma 2: Nro. de Casos de Soporte
sns.histplot(data=df_processed, x='Nro. de Casos de Soporte', kde=True, bins=20, ax=axes[1], color='lightgreen')
axes[1].set_title('Distribución de Casos de Soporte')
```

## 3) Boxplot

```python
# Boxplot de la variación de días desde la ultima conexión
plt.figure(figsize=(10, 4))
sns.boxplot(x=df_processed['Var. Días desde Ult. Conexión'])
plt.title('Variación de Días desde Ultima Conexión')
plt.xlabel('Cantidad')
plt.grid(True, axis='x')
```
