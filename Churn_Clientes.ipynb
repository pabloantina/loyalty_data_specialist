# Gr√°ifcos de int√©res

## 1) Histogramas

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo general
sns.set(style="whitegrid")

# Creo los histogramas en fila
fig, axes = plt.subplots(1, 2, figsize=(18, 5))

# Histograma 1: Meses de suscripci√≥n
sns.histplot(data=df_processed, x='Meses de suscripci√≥n', kde=True, bins=20, ax=axes[0], color='skyblue')
axes[0].set_title('Distribuci√≥n de Meses de Suscripci√≥n')

# Histograma 2: Nro. de Casos de Soporte
sns.histplot(data=df_processed, x='Nro. de Casos de Soporte', kde=True, bins=20, ax=axes[1], color='lightgreen')
axes[1].set_title('Distribuci√≥n de Casos de Soporte')
```

## 3) Boxplot

```python
# Boxplot de la variaci√≥n de d√≠as desde la ultima conexi√≥n
plt.figure(figsize=(10, 4))
sns.boxplot(x=df_processed['Var. D√≠as desde Ult. Conexi√≥n'])
plt.title('Variaci√≥n de D√≠as desde Ultima Conexi√≥n')
plt.xlabel('Cantidad')
plt.grid(True, axis='x')
```
